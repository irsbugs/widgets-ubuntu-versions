# widgets-ubuntu-versions

## Using Jupyter-notebook widgets to view Ubuntu versions

Demonstration of using the jupyter-notebook widgets: Dropdown, Label, Text and Textarea.

Ubuntu LTS releases are every two years. These iso file releases are built from over 2000 software packages. On an almost daily basis an update to one of the applications becomes available for distribution. Around every six months a new iso file is produced that incorporates the all the recent updates.

Since Ubuntu Mate was released in April 2018, there have been three new full distributions released. 

This Jupyter-notebook program will retrieve the manifests of the applications and their version numbers for the four releases of Ubuntu-Mate 18.04.

Using a drop-down widget an application may be selected. In a *Text* widget, the four versions are listed. For example the four *python3* application versions for 18.04, 18.04.1, 18.04.2,and 18.04.3 are:

`3.6.5-3,    3.6.5-3ubuntu1,    3.6.7-1~18.04,    3.6.7-1~18.04.`

The program requires on-line access to the internet.

```python
# Retrieve Ubuntu manifest files from Ubuntu web-site, process and place 
# the data in a dictionary. Use widgets to be able to select a 
# software package and view any version changes. 

import urllib.request
import ipywidgets

# Retrieve the Ubuntu 18.04 set of releases.
F0 = "ubuntu-mate-18.04-desktop-amd64.manifest"
F1 = "ubuntu-mate-18.04.1-desktop-amd64.manifest"
F2 = "ubuntu-mate-18.04.2-desktop-amd64.manifest"
F3 = "ubuntu-mate-18.04.3-desktop-amd64.manifest"

URL_1 = "http://cdimage.ubuntu.com/ubuntu-mate/releases/18.04/release/"
    
file_list = [URL_1 + F0, URL_1 + F1, URL_1 + F2, URL_1 + F3,]

# Get Ubuntu manifest files from Ubuntu web-site. 
# Process and place data in a dictionary.
application_dict = {}
for index, file in enumerate(file_list):
    with urllib.request.urlopen(file) as f:
        file_string = f.read().decode('utf-8')
        file_string = file_string[:-1] # Remove surplus newline
        file_list = file_string.split("\n")
        for line in file_list:
            line_list = line.split("\t")
            #print(line_list)

            if line_list[0] in application_dict:
                # Exists, so append
                application_dict[line_list[0]].append(line_list[1])
                
            else: 
                # New, so add product name, and pad with a ""'s if necessary
                if index == 0:
                    application_dict[line_list[0]] = [line_list[1]]                    
                if index == 1:
                    application_dict[line_list[0]] = ['x', line_list[1]]
                if index == 2:
                    application_dict[line_list[0]] = ['x', 'x', line_list[1]]
                if index == 3:
                    application_dict[line_list[0]] = ['x', 'x', 'x', line_list[1]] 
                # index = 4 is not used but allows for expansion    
                if index == 4: 
                    application_dict[line_list[0]] = ['x', 'x', 'x', 'x', line_list[1]] 
                #print(line_list[0], application_dict[line_list[0]])                            

        # Check for deprecated and pad the list with empty string:
        for key, value  in application_dict.items():
            if len(value) != index + 1:
                #print(key, value)
                application_dict[key].append('x')
                #print(key, value)
                        
#print(len(application_dict)) # # 2033      
#for key, value in sorted(application_dict.items()):
#    print("\n{}\n{}".format(key, ", ".join(value) + "."))


MESSAGE_0 = "Ubuntu LTS 18.04 releases."
MESSAGE_1 = ("Ubuntu LTS 18.04 was first released in April 2018. "
             "As of Oct 2019 there have been three update releases. \n"
             "The total software packages over the four releases is {}. \n"
             "Select a software package and the versions included in "
             "each release will be displayed."
             .format(len(application_dict)))

#=== Label widget
label = ipywidgets.Label(
    value = MESSAGE_0,
    layout = ipywidgets.Layout(width='95%',),
    style = {'description_width': 'initial'},
)

#=== Textarea Widget
text_message = ipywidgets.Textarea(
    value=MESSAGE_1,
    placeholder='Type something',
    description='',
    disabled=False,
    style={'description_width': 'initial'}, 
    layout = ipywidgets.Layout(width='95%', height='75px',)
)

#=== Text Widget
text = ipywidgets.Text(
    value='Select a product and view what versions have been release',
    placeholder='Type something',
    description='Release Versions:',
    disabled=False,
    style={'description_width': 'initial'}, 
    layout = ipywidgets.Layout(width='95%',) # height='300px',)
)

#=== Dropdown widget
key_list = sorted(application_dict.keys())

dropdown = ipywidgets.Dropdown(
    options=key_list,
    value=key_list[1],
    description='Select:',
    disabled=False,
    style = {'description_width': 'initial'},    
)

def dropdown_handle_change(names):
    # From application_dictionary use names.new as the key to get value list
    # Convert list to a comma seperated string and add a full stop.
    text.value = ",  ".join(application_dict[names.new]) + "."
    
dropdown.observe(dropdown_handle_change, names='value')
    
# Force a Dropdown event to update the Text box 
dropdown.value = key_list[0]

display(ipywidgets.VBox([label, text_message, dropdown, text,]))

# Note: Expect a few seconds delay in displaying widgets while files are retrieved.
```

# Debian package version comparison

The above code was adapted to provide a comparison between versions 8, 9 and 10 of the Debian Mate Live amd64 iso-hybrid releases. The release dates were: 

* 8.0 - April 2015
* 9.0 - June 2017
* 10.0 - July 2019

```python
# Retrieve debian manifest files from distro web-site, process and place 
# the data in a dictionary. Use widgets to be able to select a 
# software package and view any version changes. 

import urllib.request
import ipywidgets

# Retrieve Debian live releases 8, 9 and 10.
URL_1 = "https://ftp.acc.umu.se/cdimage/archive/8.0.0-live/amd64/iso-hybrid/"    
F1 = "debian-live-8.0.0-amd64-mate-desktop.iso.packages"

URL_2 = "https://ftp.acc.umu.se/cdimage/archive/9.0.0-live/amd64/iso-hybrid/"
F2= "debian-live-9.0.0-amd64-mate.packages"

URL_3 = "https://ftp.acc.umu.se/cdimage/archive/10.0.0-live/amd64/iso-hybrid/"
F3 = "debian-live-10.0.0-amd64-mate.packages"

file_list = [URL_1 + F1, URL_2 + F2, URL_3 + F3]
# Get manifest files from web-site. 
# Process and place data in a dictionary.
application_dict = {}
for index, file in enumerate(file_list):
    with urllib.request.urlopen(file) as f:
        file_string = f.read().decode('utf-8')
        file_string = file_string[:-1] # Remove surplus newline
        file_list = file_string.split("\n")
        for line in file_list:
            line_list = line.split("\t")
            # In case a tab is replaced with a space
            if len(line_list) != 2:
                line_list = line.split(" ")    
            #print(line_list)

            if line_list[0] in application_dict:
                # Exists, so append
                application_dict[line_list[0]].append(line_list[1])
                
            else: 
                # New, so add product name, and pad with a ""'s if necessary
                if index == 0:
                    application_dict[line_list[0]] = [line_list[1]]                    
                if index == 1:
                    application_dict[line_list[0]] = ['x', line_list[1]]
                if index == 2:
                    application_dict[line_list[0]] = ['x', 'x', line_list[1]]
                if index == 3:
                    application_dict[line_list[0]] = ['x', 'x', 'x', line_list[1]] 
                if index == 4:
                    application_dict[line_list[0]] = ['x', 'x', 'x', 'x', line_list[1]]                     
                #print(line_list[0], application_dict[line_list[0]])                            

        # Check for deprecated and pad the list with empty string:
        for key, value  in application_dict.items():
            if len(value) != index + 1:
                #print(key, value)
                application_dict[key].append('x')
                #print(key, value)
                        
#print(len(application_dict)) # # 2033      
#for key, value in sorted(application_dict.items()):
#    print("\n{}\n{}".format(key, ", ".join(value) + "."))

MESSAGE_0 = "Debian Mate Live 8.0, 9.0 & 10.0 releases."

MESSAGE_1 = ("Comparison of package versions for Debian Mate Live "
             "amd64 iso-hybrid releases. The release dates were: 8.0 "
             "Apr 2015, 9.0 Jun 2017, and 10 Jul 2019. \n"
             "The total software packages over the three releases is {}."
             .format(len(application_dict)))

#=== Label widget
label = ipywidgets.Label(
    value = MESSAGE_0,
    layout = ipywidgets.Layout(width='95%',),
    style = {'description_width': 'initial'},
)

#=== Textarea Widget
text_message = ipywidgets.Textarea(
    value=MESSAGE_1,
    placeholder='Type something',
    description='',
    disabled=False,
    style={'description_width': 'initial'}, 
    layout = ipywidgets.Layout(width='95%', height='75px',)
)

#=== Text Widget
text = ipywidgets.Text(
    value='Select a product and view what versions have been release',
    placeholder='Type something',
    description='Release Versions:',
    disabled=False,
    style={'description_width': 'initial'}, 
    layout = ipywidgets.Layout(width='95%',) # height='300px',)
)

#=== Dropdown widget
key_list = sorted(application_dict.keys())

dropdown = ipywidgets.Dropdown(
    options=key_list,
    value=key_list[1],
    description='Select:',
    disabled=False,
    style = {'description_width': 'initial'},    
)

def dropdown_handle_change(names):
    # From application_dictionary use names.new as the key to get value list
    # Convert list to a comma seperated string and add a full stop.
    text.value = ",  ".join(application_dict[names.new]) + "."
    
dropdown.observe(dropdown_handle_change, names='value')
    
# Force a Dropdown event to update the Text box 
dropdown.value = key_list[0]

display(ipywidgets.VBox([label, text_message, dropdown, text,]))

# Note: Expect ~10 seconds delay before displaying widgets while files are retrieved.
```
